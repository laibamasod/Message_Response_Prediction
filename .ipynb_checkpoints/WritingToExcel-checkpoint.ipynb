```python
# timestamps

import pandas as pd

# Read the existing Excel file
existing_df = pd.read_excel('MessageData.xlsx')

# Define the range of rows to update (skip the first 20 entries, update the next 50)
start_row = 162
end_row = start_row + 35

# Read the new timestamps from the 'conversations.txt' file
with open('conversations.txt', 'r') as file:
    data = [line.strip().split(' | ') for line in file]

# Extract the timestamps for the next 50 entries
timestamps_data = data[:35]

# Create a DataFrame for the new timestamps
timestamps_df = pd.DataFrame(timestamps_data, columns=['TimeStamp1', 'TimeStamp2'])

# Update the 'TimeStamp1' and 'TimeStamp2' columns for the specified rows
existing_df.loc[start_row:end_row - 1, ['TimeStamp1', 'TimeStamp2']] = timestamps_df.values

# Write the updated data back to the Excel file
existing_df.to_excel('MessageData.xlsx', index=False)
```

```python
# TimeDelay 

import pandas as pd
from datetime import datetime

# Read the existing Excel file
df = pd.read_excel('MessageData.xlsx')

# Convert the 'TimeStamp1' and 'TimeStamp2' columns to datetime objects
df['TimeStamp1'] = pd.to_datetime(df['TimeStamp1'])
df['TimeStamp2'] = pd.to_datetime(df['TimeStamp2'])

# Calculate the time delay and add it as a new column 'TimeDelay' in seconds
df['TimeDelay'] = (df['TimeStamp2'] - df['TimeStamp1']).dt.total_seconds()

# Reorder the columns to have 'TimeDelay' at the end
df = df[['Message', 'Response', 'isResponse', 'TimeStamp1', 'TimeStamp2', 'TimeDelay','Common_Keyword_Count', 'Has_Questions']]

# Write the updated data back to the Excel file
df.to_excel('MessageData.xlsx', index=False)

```

```python
# common keyword 


import pandas as pd

# Load the Excel data into a Pandas DataFrame
df = pd.read_excel('MessageData.xlsx')

# Tokenize the 'Message' and 'Response' columns, and convert to lowercase for case-insensitive comparison
df['Message'] = df['Message'].str.lower().str.split()
df['Response'] = df['Response'].str.lower().str.split()

# Find common keywords for each row
df['Common_Keywords'] = df.apply(lambda row: set(row['Message']) & set(row['Response']), axis=1)

# Count and analyze the common keywords
df['Common_Keyword_Count'] = df['Common_Keywords'].apply(len)

# Reorder the columns to have 'TimeDelay' at the end
df = df[['Message', 'Response', 'isResponse', 'TimeStamp1', 'TimeStamp2', 'TimeDelay', 'Common_Keyword_Count']]

# Write the updated data back to the Excel file
df.to_excel('MessageData.xlsx', index=False)

```

```python
# Questions

import pandas as pd

# Read the Excel file
df = pd.read_excel('MessageData.xlsx')

# Function to check for questions and 'yes' or 'no' in Message and Response
def has_questions_and_yesno(row):
    message_has_question = '?' in str(row['Message'])
    response_has_yesno = any(word.lower() in ['yes', 'no', 'yes,', 'no,'] for word in str(row['Response']).lower().split())
    print(response_has_yesno)
    return message_has_question and response_has_yesno



# Apply the function to each row and create a new 'Has_Questions' column
df['Has_Questions'] = df.apply(has_questions_and_yesno, axis=1)

# Write the updated DataFrame back to the Excel file
df.to_excel('MessageData.xlsx', index=False)
```
